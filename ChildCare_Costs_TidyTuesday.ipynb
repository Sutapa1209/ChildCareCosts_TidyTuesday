```r
# Let's visualize the trends in Center-Based Care Costs for School-Age Children across US counties from 2008 to 2018.

childcare_costs %>%
  filter(!is.na(mcsa)) %>%
  ggplot(aes(x = study_year, y = mcsa, group = study_year, fill = study_year)) +
  geom_boxplot(alpha = 0.8, show.legend = FALSE) +
  scale_fill_distiller(palette = "Blues") +
  labs(
    title = "Trends in Center-based Care for School Age Children Over Time",
    x = "Year",
    y = "Weekly School-Age Child Care Cost"
  ) +
  theme_minimal()
```

```r
#Over this 10-year span, there's a clear upward trend in median costs, suggesting that childcare for school-age children has gradually become more expensive nationwide.
#Each box represents the distribution of costs in a given year, and we observe that not only are the medians increasing, but the spread (interquartile range) is also widening over time.
#This indicates growing variability in what families are paying, with some counties facing disproportionately higher costs in more recent years.
#The consistent rise in both cost and variability highlights increasing financial pressure on families seeking care for older children.
```

```r
# Let's plot the national trend in median household income (adjusted to 2018 dollars) from 2008 to 2018.

childcare_costs %>%
  filter(!is.na(mhi_2018)) %>%
  group_by(study_year) %>%
  summarise(median_income = median(mhi_2018, na.rm = TRUE)) %>%
  ggplot(aes(x = study_year, y = median_income)) +
  geom_line(color = "steelblue", size = 1.2) +
  geom_point(color = "steelblue", size = 2) +
  labs(
    title = "National Trend of Median Household Income (2008-2018)",
    x = "Year",
    y = "Median Household Income (2018 Dollars)"
  ) +
  theme_minimal()
```

```r
#The overall pattern reveals a dip in income following the 2008 financial crisis, with a notable decline between 2010 and 2014.
#However, after reaching a low point around 2014, household incomes began to recover, showing a steady upward trend from 2015 to 2018.
#By the end of the period, median income levels surpassed pre-recession figures, reflecting broader economic recovery during these years.
```

```r
# Let's compare the weekly cost of center-based care for different age groups- infants, toddlers, and preschoolers over time.

# First, we'll reshape the dataset from wide to long format to make plotting easier in ggplot2

library(tidyverse)

# Reshape to long format
cost_long <- childcare_merged %>%
  select(study_year, mc_infant, mc_toddler, mc_preschool) %>%
  pivot_longer(
    cols = starts_with("mc_"),
    names_to = "age_group",
    values_to = "weekly_cost"
  ) %>%
  drop_na()

cost_long

# this table has one row per county per type. For example, in 2010, for example, we have dozens or hundreds of counties
#And for each one, we'll have weekly cost data for: mc_infant, mc_toddler, mc_preschool
```

```r
# For getting the national average cost for each age group, per year-- we calculate the average weekly cost across all those counties

avg_costs <- cost_long %>%
  group_by(study_year, age_group) %>%
  summarise(mean_cost = mean(weekly_cost, na.rm = TRUE), .groups = "drop")

# Now let's plot it using ggplot!
ggplot(avg_costs, aes(x = study_year, y = mean_cost, color = age_group)) +
  geom_line(size = 1.2) +
  geom_point() +
  labs(
    title = "Center-Based Childcare Costs Over Time by Age Group",
    x = "Study Year",
    y = "Average Weekly Cost (USD)",
    color = "Age Group"
  ) +
  scale_color_manual(
    values = c("mc_infant" = "darkred", "mc_toddler" = "orange", "mc_preschool" = "darkgreen"),
    labels = c("mc_infant" = "Infant", "mc_toddler" = "Toddler", "mc_preschool" = "Preschool")
  ) +
    theme_minimal(base_size = 12)
```

```r
# All 3 types of care are ricing overtime, with infant care consistently being the most expensive form of care — likely due to lower caregiver-to-child ratios required
# Toddler care is in the middle, and rising steadily.
# Preschool care is the least expensive, but still rising over time.
# There’s a notable dip in 2016 — worth exploring (possibly due to data quality, policy change, economic factor).
```

```r
# SECTION III: How Gender dynamics, Labor Force Participation Rates interact and ultimately affect Infant care costs.

#Let's visualise how the Female labor force participation rate (for women aged 20–64) at the county level relates to The weekly cost of center-based infant care in those counties.
# Plotting Graph of y axis= Infant Care Cost vs x axis= Female Labor Force Participation
# Each point represents one county-year observation.
# The black line is a linear regression fit that helps illustrate the overall trend.

childcare_merged %>%
  ggplot(aes(x = flfpr_20to64, y = mc_infant)) +
  geom_point(alpha = 0.4, color = "#FFB6C1") +
  geom_smooth(method = "lm", se = FALSE, color = "black") +
  labs(
    title = "Infant Care Cost vs Female Labor Force Participation",
    x = "Female Labor Force Participation (Age 20–64)",
    y = "Weekly Infant Care Cost"
  ) +
  theme_minimal()

cor.test(childcare_merged$flfpr_20to64, childcare_merged$mc_infant, use = "complete.obs")
```

```r
#There’s a slight-moderate positive correlation between female labor force participation and infant care costs.
#As the percentage of women participating in the labor force increases, so does the cost of infant care.
#Counties with higher female workforce engagement tend to spend more on infant care.
#However, there is a wide spread of costs even at similar participation levels—suggesting other factors (like income, region, policy) may also play a role.
```

```r
# Thus, we can form a hypothesis:
# Higher demand (i.e., more women working) could potentially predict the price of childcare services like infant care.
# We can use a multivariable regression model to test whether this holds after controlling for income, urbanicity, race, etc.

# Including all race variables (like percent White, Black, Asian, etc.) in a regression causes multicollinearity, because these percentages add up to 100%.
#This would make the model unstable and the coefficients uninterpretable.
#So, we typically leave out one group (like White) as a reference, and interpret the other race coefficients relative to that group (Black and Hispanic in this case).
```

```r
# checking the skewness of infant care costs

ggplot(childcare_merged, aes(x = mc_infant)) +
  geom_histogram(fill = "#5F9ED1", color = "white", bins = 30) +
  labs(
    title = "Distribution of Weekly Infant Care Costs",
    x = "Weekly Infant Care Cost (USD)",
    y = "Number of Counties"
  ) +
  theme_minimal()

# Since infant care costs are right-skewed (most counties have lower infant care cost around $100),
#we will use log transformation for mc_infant before putting it in a regression model
```

```r
# Running the regression model with female labor force participant rate (IV) and log transformed infant care costs (DV)
childcare_merged <- childcare_merged %>%
  mutate(log_mc_infant = log(mc_infant + 1))  # Add 1 to avoid log(0)

model_log <- lm(log_mc_infant ~ flfpr_20to64 + mhi_2018 + one_race_b + hispanic, data = childcare_merged)

summary(model_log)
```

```r
# takeaways:
#The model explains approximately 35.5% of the variance in log-transformed infant care costs (Adjusted R² = 0.3551), with all predictors statistically significant (p < 0.05).
#Results indicate that higher female labor force participation and higher median household income are both associated with increased infant care costs.
#Additionally, counties with larger Hispanic populations tend to have slightly higher costs, while those with larger Black-only populations tend to have slightly lower costs (with white populations as reference).
#Because the outcome variable was log-transformed, these coefficients reflect approximate percentage changes in infant care costs associated with each predictor, holding other variables constant.
```

```r
#Comparing Male vs Female Labor Force Participation Rates on Infant Care Cost (INTERACTION TERM OF GENDER!)
library(tidyr)
library(dplyr)

#INTERACTION

# Prepare data
gender_model_data <- childcare_merged %>%
  filter(!is.na(flfpr_20to64), !is.na(mlfpr_20to64), !is.na(log_mc_infant)) %>%
  select(log_mc_infant, flfpr_20to64, mlfpr_20to64) %>%
  pivot_longer(cols = c(flfpr_20to64, mlfpr_20to64),
               names_to = "gender", values_to = "lfpr") %>%
  mutate(gender = recode(gender,
                         "flfpr_20to64" = "Female",
                         "mlfpr_20to64" = "Male"))

# Model with interaction
interaction_model <- lm(log_mc_infant ~ lfpr * gender, data = gender_model_data)
summary(interaction_model)
```

```r
#INTERESTING RESULTS!
#There is a significant relationship between labor force participation rate (LFPR) and infant care costs, and this relationship differs by gender.
#Specifically, counties with higher female LFPR are associated with sharper increases in infant care costs than counties with higher male LFPR.
#This difference is statistically significant (p < 2e-16), suggesting that female labor force participation may drive demand for infant care more strongly than male participation.
```

```r
#This provides empirical support our earlier hypothesis:
#infant care markets may be more responsive to women’s labor force participation (who traditionally bear more caregiving responsibilities), than males.
```

```r
# SECTION IV: Racial/ethnic differences, Family Structure, and Occupational category impact on Indant Childcare Costs

#Let's run a multiple linear regression model predicting infant childcare costs (mc_infant) from racial composition, income, and unemployment rates with White as the reference group:
model_race_income_unemp <- lm(mc_infant ~ one_race_a + one_race_b + one_race_h +
                                          mhi_2018 + unr_20to64,
                               data = childcare_costs)

summary(model_race_income_unemp)
```

```r
#This model shows race and ethnicity matters a lot in predicting infant child care costs relative to white groups
#Counties with higher Asian population shares are strongly and positively associated with higher infant care costs — even after controlling for income and unemployment.
#Black and Hispanic population shares are associated with slightly lower infant care costs.
# Income remains a strong predictor — higher-income counties pay more.
# Interestingly, higher unemployment is also associated with higher costs. However, given that the majority of the sample is White, the observed relationship might not generalize equally across racially diverse counties,
#and further subgroup analysis (e.g., by race or region) could clarify these dynamics.
```

```r
#Plotting counties with the highest concentration of specific racial or ethnic groups could offer useful insights into how racial composition relates to childcare costs.
#However, it’s important to note that such an analysis may oversimplify population diversity in many areas
#particularly in states with high racial/ethnic heterogeneity — and may not fully capture the nuance of multiethnic counties.
```

```r
# Assinging each county based on the highest race proportion:

childcare_race_plot <- childcare_merged %>%
  filter(!is.na(mc_infant)) %>%
  mutate(
    race_group = case_when(
      one_race_w == pmax(one_race_w, one_race_b, one_race_a, hispanic, na.rm = TRUE) ~ "White",
      one_race_b == pmax(one_race_w, one_race_b, one_race_a, hispanic, na.rm = TRUE) ~ "Black",
      hispanic == pmax(one_race_w, one_race_b, one_race_a, hispanic, na.rm = TRUE) ~ "Hispanic",
      one_race_a == pmax(one_race_w, one_race_b, one_race_a, hispanic, na.rm = TRUE) ~ "Asian",
      TRUE ~ "Other"
    )
  )

#Calculate average infant care cost per group

avg_race_cost <- childcare_race_plot %>%
  group_by(race_group) %>%
  summarise(mean_infant_cost = mean(mc_infant, na.rm = TRUE))

#Plotting!
ggplot(avg_race_cost, aes(x = race_group, y = mean_infant_cost, fill = race_group)) +
  geom_bar(stat = "identity", width = 0.7, alpha = 0.9) +
  labs(
    title = "Average Infant Care Costs by Dominant Racial Group in County",
    x = "Dominant Racial Group",
    y = "Weekly Infant Care Cost (USD)"
  ) +
  scale_fill_manual(values = c(
    "White" = "#FFD1DC",
    "Black" = "#7EA8BE",
    "Hispanic" = "#F4A261",
    "Asian" = "#A0CED9",
    "Other" = "#BDB2FF"
  )) +
  theme_minimal(base_size = 13)


```

```r
# The plot confirm our previous findings from the regression model
# Counties with an Asian-majority population have significantly higher infant care costs (~$280/week), standing out from all other racial groups.
# An xgboost (machine learning) model analysis of the same dataset by Julia Silge, found similar patterns.
# Asian-majority counties may be more urban, affluent, and located in high-cost-of-living areas, driving up childcare prices.
# Whereas White-majority counties have moderately high costs (~$140–150/week), and Black- and Hispanic-majority counties might be concentrated in areas with lower median incomes or less expensive childcare infrastructure.
```

```r
# Occupation Type and School-Age Childcare Costs by County
# Let's explore if there is a relationship between the percentage of county-level employment in different occupational sectors and the weekly cost of school-age childcare (mcsa)
# Each point represents a U.S. county, and color indicates median household income (lighter shades = higher income).

library(tidyverse)

# Define readable labels for each occupation type
occupation_labels <- c(
  emp_m = "Business,Science,Arts",
  emp_service = "Service Occupations",
  emp_sales = "Sales & Office",
  emp_n = "Construction & Maintenance",
  emp_p = "Production & Transportation"
)

# Create the plot
childcare_costs |>
  select(mcsa, starts_with("emp_"), mhi_2018) |>
  pivot_longer(
    cols = starts_with("emp_"),
    names_to = "occupation_type",
    values_to = "percent_employed"
  ) |>
  ggplot(aes(x = percent_employed, y = mcsa, color = mhi_2018)) +
  geom_point(alpha = 0.5) +
  facet_wrap(
    ~ occupation_type,
    scales = "free_x",
    nrow = 2,
    labeller = as_labeller(occupation_labels)
  ) +
  scale_color_viridis_c() +
  labs(
    x = "% Employed in Occupation",
    y = "Weekly School-Age Childcare Cost",
    color = "Median Income",
    title = "Occupation Type and School-Age Childcare Costs by County"
  ) +
  theme_minimal()
```
